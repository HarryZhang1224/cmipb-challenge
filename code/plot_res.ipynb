```r
library(ggplot2)
library(ggsci)
library(dplyr)
```

```r
res_df <- read.csv("~/cmi-challenge/data/pred_results.csv")
```

```r
res_df
```

```r
options(repr.plot.width = 8, repr.plot.height = 3.5, repr.plot.res = 500)
res_df %>%
    group_by(target, task) %>%
    summarize(mean=mean(cor, na.rm=TRUE), sd=sd(cor, na.rm=TRUE)) %>%
    ggplot(., aes(target, mean, fill=task)) +
        geom_bar(stat="identity", color="black", width=0.5) +
        geom_errorbar(aes(ymin=mean-sd, ymax=mean+sd), width=0.2, position=position_dodge(0.9)) +
        theme_bw() +
        scale_fill_observable() +
        ylab("Mean Spearman's R") +
        xlab("") +
        labs(fill="")
```
